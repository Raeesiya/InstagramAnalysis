# Instagram Reach Analysis

Instagram is one of the most popular social media applications today. People using Instagram professionally are using it for promoting their business, building a portfolio, blogging, and creating various kinds of content. As Instagram is a popular application used by millions of people with different niches, Instagram keeps changing to make itself better for the content creators and the users. But as this keeps changing, it affects the reach of our posts that affects us in the long run. So, if a content creator wants to do well on Instagram in the long run, they have to look at the data of their Instagram reach. That is where the use of Data Science comes in
Below are all the features in the data:

1.	Impressions: Number of impressions in a post (Reach)
2.	From Home: Reach from home
3.	From Hashtags: Reach from Hashtags
4.	From Explore: Reach from Explore
5.	From Other: Reach from other sources
6.	Saves: Number of saves
7.	Comments: Number of comments
8.	Shares: Number of shares
9.	Likes: Number of Likes
10.	Profile Visits: Number of profile visits from the post
11.	Follows: Number of Follows from the post
12.	Caption: Caption of the post
13.	Hashtags: Hashtags used in the post

## visualization

#### Heatmap

```python
# Correlation heatmap for numerical columns
numerical_data = df.select_dtypes(include=['int64'])
plt.figure(figsize=(20, 8))
sns.heatmap(numerical_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap', fontsize=16)
plt.show()

```

#### Bar graph

```python
# Assume the CSV has columns 'Likes' and 'Comments'
likes = df['Likes']
comments = df['Comments']

# Create a bar chart for likes and comments
plt.figure(figsize=(10, 5))
bar_width = 0.35
index = range(len(df))

plt.bar(index, likes, bar_width, label='Likes', color='blue', alpha=0.6)
plt.bar([i + bar_width for i in index], comments, bar_width, label='Comments', color='red', alpha=0.6)

# Add titles and labels
plt.title('Comparison of Likes and Comments')
plt.xlabel('Posts')
plt.ylabel('Count')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

#### scatterplot

```python
import seaborn as sns

# Scatter plot for Likes vs Impressions
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Impressions', y='Likes', data=df, color='purple', s=100)
plt.title('Likes vs Impressions', fontsize=16)
plt.xlabel('Impressions', fontsize=12)
plt.ylabel('Likes', fontsize=12)
plt.show()
```

#### Pie Chart

```python
# Summing the sources of impressions
sources = ['From Home', 'From Hashtags', 'From Explore', 'From Other']
impression_sums = df[sources].sum()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(impression_sums, labels=sources, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 'orange'])
plt.title('Proportion of Impressions Sources', fontsize=16)
plt.show()
```

#### WordCloud

```python
from wordcloud import WordCloud

# Combine all hashtags
all_hashtags = ' '.join(df['Hashtags'])

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_hashtags)

# Display word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Common Hashtags', fontsize=16)
plt.show()
```
